#### Renaming Columns

```python
turnover_df = pd.read_csv("gh_turnover_four_projects.csv")

turnover_df = (
    turnover_df.rename(columns={
        "domain": "project_domain",
        "language": "project_language",
        "windows": "project_age",
        "window_idx": "quarter",
        "num_team": "team_size",
        "num_commits": "project_commits",
        "blau_gender": "gender_blau",
        "Gini_gh_ten": "gh_tenure_gini",
        "Core1": "core_dev",
        "commits": "user_commits",
        "propCommits": "commit_proportion",
        "github_tenure": "gh_tenure",
        "LeavesNextQ": "leaves_next_q"
    })
)
```

#### Calculate the mean and median of `user_commits`

```python
mean_commits = turnover_df["user_commits"].mean()
median_commits = turnover_df["user_commits"].median()

print("Mean:", mean_commits)
print("Median:", median_commits)
```

#### Select all columns that start with a *p* (i.e., `(^p)`) **or** contain an *g* (i.e., `(g)`). Save it as a new DataFrame named `turnover_new`. Output the columns in your notebook

```python
turnover_new = turnover_df.filter(regex=r'^(p|.*g)')
turnover_new.head()
```

#### Using your newly created DataFrame `turnover_new`, select rows in which `project_commits` is greater than 9,000

```python
turnover_new = (
    turnover_new.assign(project_commits=turnover_df["project_commits"])  # Add for filtering if missing
    .query("project_commits > 9000")
)
turnover_new.head()
```

#### Using pandas piping notation, perform the previous two operations together and save it as a new DataFrame `turnover_newer`. This means you should select columns that start with a *p* or contain an *g* and select rows in which `project_commits` is greater than 9,000.

```python
turnover_newer = (
    turnover_df
    .filter(regex=r'^(p|.*n)')
    .assign(project_commits=turnover_df["project_commits"])  # For filtering again
    .query("project_commits > 9000")
)
turnover_newer.head()
```

#### Using pandas piping notation, select all columns that end with the letter *e* (i.e., `e$`) or contain the letter *s* (i.e., `s`). Additionally, select rows in which `gh_tenure` is greater than or equal to 1,991

```python
filtered_df = (
    turnover_df
    .filter(regex=r'(e|s)')
    .query("gh_tenure > 1.99")
)
filtered_df.head()
```

#### Use the query you just performed. You will calculate the mean and median of `user_commits`. How does it compare to the answers above in which you calculated the mean and median? Please provide your answer in a Markdown cell within your notebook

```python
mean_filtered = filtered_df["user_commits"].mean()
median_filtered = filtered_df["user_commits"].median()

print("Filtered Mean:", mean_filtered)
print("Filtered Median:", median_filtered)
```
